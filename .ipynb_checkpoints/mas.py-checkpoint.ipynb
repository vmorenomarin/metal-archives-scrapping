
# Abreviations 
- `bn`: band name

- `th`: lyrical theme
- `la`: label

```python
category = menu()
id = input("Write word for {0}: ".format(category))

if ' ' in id:
    id=id.replace(' ',"_")
    
ids = (id,category)
url = "https://www.metal-archives.com/{1}/{0}".format(ids[0],ids[1])

site = urllib.request.urlopen(url).read()
content = bs4.BeautifulSoup(site)
content_p = content.prettify

```

```python
#Band statistics

bandId=content.findAll('script')[4].text.split("bandId = ")[1].split(";")[0]
bn = content.div.h1.text

stats=[bn]
for i in range(8):
    item=content.findAll('dd')[i].text.replace("\t","").replace("\n","").strip()
    stats.append(item)
comment=content.find('div',{'class':'band_comment'}).text.replace('\n',"").replace('\t',"").replace('\r',"").strip()    

stats.append(comment)
file_gen(id,content_p)


cols=["Band name","Country","Location","Status","Formed in","Genre","Lyrical themes","Curret label","Years active","Comment"]
info=pd.DataFrame([stats],index=["Info"],columns=cols)
pd.set_option('display.max_colwidth', -1)
info

```

```python
### Band discography

url_disc='https://www.metal-archives.com/band/discography/id/{0}/tab/all'.format(bandId)
discog=urllib.request.urlopen(url_disc).read()

disc_content=bs4.BeautifulSoup(discog)

list_album=disc_content.tbody.findAll('tr')
number_album=len(list_album)

list_album=[]

for i in range(number_album):
    disc=disc_content.tbody.findAll('tr')[i]
    album=[]
    for j in range(0,3):
        album.append(disc.findAll('td')[j].text)
    list_album.append(album)

cols=['Name','Type','Year']    
discography = pd.DataFrame(list_album, index=range(1,number_album+1), columns=cols)
pd.set_option('display.max_colwidth', -1)

discography
```

```python
# Band members

#currents
number_currents=len(content.find(id='band_tab_members_current').findAll('tr', {"class":'lineupRow'}))


current_list=[]
for i in range(number_currents):
    item=content.find(id='band_tab_members_current').findAll('tr', {"class":"lineupRow"})[i]
    member=[]
    for j in range(0,2):
        member.append(item.findAll('td')[j].text.replace("\n","").replace("\t","").replace("\xa0"," "))
    
    a=item.find_next_sibling('tr').text
    if re.search("See also:",a) :
            member.append(item.findNext('tr', {"class":'lineupBandsRow'}).text.replace("\n","").replace("\t","").replace("\xa0"," "))
    else:
        member.append("")    
    current_list.append(member)

pd.set_option('display.max_colwidth', -1)
cols=["Name (or alias)", "Band rol", "Other bands"]
current_lineup=pd.DataFrame(current_list, index=range(1,number_currents+1), columns=cols)
current_lineup
```

```python
#past
number_past=len(content.find(id='band_tab_members_past').findAll('tr', {"class":'lineupRow'}))

past_list=[]
for i in range(number_past):
    item=content.find(id='band_tab_members_past').findAll('tr', {"class":"lineupRow"})[i]
    member=[]
    for j in range(0,2):
        member.append(item.findAll('td')[j].text.replace("\n","").replace("\t","").replace("\xa0"," "))
    
    #a=item.find_next_sibling('tr').text
    if item.find_next_sibling('tr') and re.search("See also:",item.find_next_sibling('tr').text) :
            member.append(item.findNext('tr', {"class":'lineupBandsRow'}).text.replace("\n","").replace("\t","").replace("\xa0"," "))
    else:
        member.append("")    
    past_list.append(member)
    
past_lineup=pd.DataFrame(past_list, index=range(1,number_past+1), columns=cols)
past_lineup    
```

```python
import bs4
import urllib
import pandas as pd
import numpy as np
from sys import exit
import os
from funcs import *
from class_ma import *
import warnings 
import re
from selenium import webdriver
import time 



browser = webdriver.Chrome('./chromedriver')
#time.sleep(5)
url='https://www.metal-archives.com/search?searchString=king&type=band_name'
browser.get(url)

python_button = browser.find_element_by_id('content_wrapper')
python_button.click()
content=bs4.BeautifulSoup(browser.page_source)


number_tabs=len(content.find('div',{'class':"dataTables_paginate paging_full_numbers"}).find('span').findAll('a'))
final_list=[]
total_number_bands=0
for k in range(number_tabs):
    list_bands=[]
    button_page=browser.find_element_by_css_selector("a.next.paginate_button")
    button_page.click() 
    python_button.click()
    content=bs4.BeautifulSoup(browser.page_source)
    number_bands=len(content.tbody.find_all('tr'))
    total_number_bands+=number_bands
    links=[]
    
    for i in range(number_bands):
        band = []
        
        for j in range(3):
            item = content.tbody.find_all('tr')[i].find_all('td')[j].text
            band.append(item)
        link = content.tbody.find_all('tr')[i].td.a.get('href')    
        band.append(link)
        list_bands.append(band)
        
    final_list += list_bands
browser.close()
    
col_item=["Name","Genre","County","Link"]
bands=pd.DataFrame(final_list,index=range(1,total_number_bands+1),columns=col_item)
bands[['Name',"Genre","County"]]




    
```
